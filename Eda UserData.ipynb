```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('User_data.csv')
```

```python
data = data.drop_duplicates(keep = 'first')
```

```python
fig_size = (12, 6)
sns.set(rc={'figure.figsize':fig_size})
```

```python
sns.histplot(data['Age'], bins = 50);
```

```python
data['Gender'].value_counts().plot.bar(ylabel = 'count', rot = 45);
```

```python
data.groupby('Country').apply(lambda x : len(x)).plot.bar(ylabel = 'count', title = 'User count by country', rot = 45);
```

```python
data['Registered Year'].value_counts().plot.bar(ylabel = '$\log (count)$', title = 'User count by registered year', logy = True, rot = 45);
```

```python
data.groupby('Country')['Age'].mean().plot.bar(title = 'Mean age by country', y = 'Age', rot = 45); 
```

```python
data_pivot = data.pivot_table(index=["Country"],columns=["Gender"], values="Age", aggfunc="count")
```

```python
data_pivot.plot.bar(rot=45, title = 'Gender breakdown by country', ylabel = 'count');
```

```python
sns.heatmap(data_pivot, cmap='YlGnBu');
```

```python
data_pivot_reg_year = data.pivot_table(index=["Registered Year"],columns=["Gender"], values="Age", aggfunc="count")
```

```python
data_pivot_reg_year.plot.bar(rot=45, title = 'Gender breakdown by registered year', ylabel = 'count');
```

```python
sns.heatmap(data_pivot_reg_year, cmap='YlGnBu');
```

```python
corr = data.corr()
sns.set_theme(style = 'white')
sns.heatmap(corr, cmap = 'Blues');
```

```python
sns.boxplot(data = data['Age']);
```
